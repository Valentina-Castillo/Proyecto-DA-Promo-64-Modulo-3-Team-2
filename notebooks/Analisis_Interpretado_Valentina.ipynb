# **RESUMEN EJECUTIVO**

**Situación actual:** ABC Corporation enfrenta una tasa de rotación del 16.1% (237 empleados de 1470 se han ido). Aunque esto puede parecer moderado, el análisis revela patrones preocupantes que requieren atención inmediata.

**Hallazgo principal:** El tiempo extra (overtime) es el factor más crítico de rotación. Los empleados que trabajan tiempo extra tienen 2.3 veces más probabilidad de irse que aquellos que no lo hacen.

**Empleados en riesgo:** De los 1233 empleados actuales, 98 (7.9%) están en alto riesgo de rotación. Adicionalmente, 583 empleados (47.3%) presentan riesgo medio.

# **1\. CONTEXTO DEL PROYECTO**

En el entorno empresarial altamente competitivo de hoy en día, la toma de decisiones informadas es esencial para el éxito a largo plazo. La retención de empleados y la satisfacción en el trabajo son cuestiones críticas para cualquier organización, ya que afectan directamente a la productividad, la moral y la rentabilidad.

ABC Corporation nos ha contratado con el objetivo de reducir la rotación de empleados y mejorar la satisfacción en el trabajo. Nuestra misión es identificar factores clave que influyen en la satisfacción laboral y, en última instancia, en la retención de empleados.

# **2\. HALLAZGOS PRINCIPALES: ¿POR QUÉ SE VAN LOS EMPLEADOS?**

Nuestro análisis ha identificado cinco factores principales que explican por qué los empleados abandonan ABC Corporation:

## **2.1. TIEMPO EXTRA (OVERTIME) \- FACTOR CRÍTICO**

**⚠️ Este es el factor más importante identificado.** Los empleados que trabajan tiempo extra tienen una tasa de rotación del 30.9%, comparado con solo 10.5% para aquellos que no trabajan tiempo extra. Esto representa un aumento de 2.9x en la probabilidad de rotación.

Interpretación: El tiempo extra excesivo genera agotamiento y desequilibrio entre vida personal y trabajo, llevando a los empleados a buscar oportunidades con mejor balance.

## **2.2. BALANCE VIDA-TRABAJO**

Los empleados muy insatisfechos con su balance vida-trabajo tienen 31.2% de rotación, casi el doble que aquellos satisfechos (17.6%).

Interpretación: La falta de balance entre vida personal y profesional es un predictor fuerte de rotación. Este factor está estrechamente relacionado con el tiempo extra.

## **2.3. ROLES DE ALTO RIESGO**

Los roles con mayor rotación son:

* **Sales Representative: 39.8% de rotación**  
* **Laboratory Technician: 23.9% de rotación**  
* **Human Resources: 23.1% de rotación**

Interpretación: Sales Representative tiene una rotación alarmante del 39.8%. Este rol típicamente combina alta presión, trabajo por objetivos y posiblemente tiempo extra, lo que explica la alta tasa de salida.

## **2.4. SATISFACCIÓN LABORAL Y AMBIENTE**

Empleados totalmente insatisfechos con su trabajo: 22.6% de rotación

Empleados totalmente insatisfechos con el ambiente: 25.4% de rotación

Interpretación: Aunque importante, la insatisfacción por sí sola no es tan determinante como el tiempo extra. Sin embargo, cuando se combina con otros factores, aumenta significativamente el riesgo de rotación.

## **2.5. PERFIL DEMOGRÁFICO DE RIESGO**

Empleados solteros muestran mayor movilidad:

* Solteros: 25.7% de rotación  
* Casados: 12.7% de rotación

Interpretación: Los empleados solteros tienen mayor flexibilidad para cambiar de trabajo y menos ataduras familiares que los anclen a una ubicación o empresa específica.

## **2.6. ANÁLISIS DE COMPENSACIÓN**

**Hallazgo importante:** Los empleados que se fueron ganaban una mediana de $3348 mensuales, mientras que los que permanecen ganan $5155. Diferencia: $1807

Interpretación: Aunque la compensación es un factor, la diferencia no es dramática. Esto confirma que el problema principal es el balance vida-trabajo y las condiciones laborales, no solo el salario.

# **4\. RECOMENDACIONES Y PLAN DE ACCIÓN**

Basándonos en el análisis de datos, presentamos las siguientes recomendaciones priorizadas:

## **4.1. PRIORIDAD CRÍTICA: Gestión del Tiempo Extra**

**Acciones Recomendadas:**

1. Implementar política de "No Overtime Obligatorio"  
2. Auditar departamentos con alto overtime  
3. Contratar personal adicional en áreas críticas  
4. Establecer límites de horas extras (máximo 10 horas/semana)

**Impacto Esperado:** Reducción estimada de 30-40% en la tasa de rotación.

## **4.2. PRIORIDAD ALTA: Programa de Balance Vida-Trabajo**

**Acciones Recomendadas:**

5. Implementar trabajo flexible y remoto  
6. Crear política de desconexión fuera de horario  
7. Encuestas trimestrales de balance  
8. Beneficios de bienestar

**Impacto Esperado:** Reducción de 20-25% en rotación.

## **4.3. PRIORIDAD ALTA: Intervención en Sales**

**Acciones Recomendadas:**

9. Revisión de estructura de compensación  
10. Programa de mentoría  
11. Definir metas realistas

**Impacto Esperado:** Reducir rotación en Sales de 39.8% a \~25%.

## **4.4. PRIORIDAD MEDIA: Estrategia de Retención Proactiva**

**Acciones Recomendadas:**

12. Conversaciones uno-a-uno con empleados en riesgo  
13. Planes de desarrollo personalizado  
14. Programa de reconocimiento  
15. "Stay interviews" regulares

# **5\. PLAN DE IMPLEMENTACIÓN**

## **Fase 1: Acciones Inmediatas (30 días)**

## **Fase 2: Cambios Estructurales (60-90 días)**

# **6\. CONCLUSIONES**

El análisis de datos de ABC Corporation revela que la rotación de empleados no es un problema de compensación principalmente, sino de condiciones laborales y balance vida-trabajo.

**Los tres insights más importantes son:**

16. El tiempo extra es el factor más crítico  
17. Sales necesita atención urgente (39.8% de rotación)
